```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
file_path = '../data/breast_cancer.csv' 
data = pd.read_csv(file_path)
```

```python
print("Dataset shape:", data.shape)
print("First few rows of the dataset:")
display(data.head())
```

```python
print("Missing values in the dataset:")
print(data.isnull().sum())
```

```python
data = data.drop(columns=['Unnamed: 32'], errors='ignore')
```

```python
data['diagnosis'].value_counts()
```

```python
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})
```

```python
print("Statistical summary:")
display(data.describe())
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=data['diagnosis'])
plt.title("Distribution of Diagnosis (0 = Benign, 1 = Malignant)")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=False, cmap='coolwarm', cbar=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# check for the how many bins can be discretized in each column
for col in data.columns:
    print(col, len(data[col].unique()))
    
```

```python
from sklearn.preprocessing import KBinsDiscretizer

# Select the numeric features (exclude 'id' and 'diagnosis')
features = data.drop(columns=['id', 'diagnosis'])

# Initialize the discretizer
n_bins = 3  # Number of bins
discretizer = KBinsDiscretizer(n_bins=n_bins, encode='ordinal', strategy='quantile')

# Fit and transform the features
discretized_features = discretizer.fit_transform(features)

# Convert the discretized data back to a DataFrame
discretized_df = pd.DataFrame(discretized_features, columns=[f"{col}" for col in features.columns])

# Combine with the target variable
data_discretized = pd.concat([data[['id', 'diagnosis']], discretized_df], axis=1)
```

```python
bin_data = {}
for col, edges in zip(features.columns, discretizer.bin_edges_):
    bin_data[col] = edges.tolist()

with open("cancer_bins.json", "w") as f:
    json.dump(bin_data, f, indent=4)
```
